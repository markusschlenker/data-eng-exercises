```python
from sqlalchemy import create_engine,text
import psycopg2
```

```python
username = "postgres"     
password = "postgres"     
host = "localhost"            
port = "5433"                   
database = "db"

NAME_TBL_NORTHWIND = "northwindtbl"

# Create the SQLAlchemy engine
engine = create_engine(f"postgresql://{username}:{password}@{host}:{port}/{database}")

# Test the connection
with engine.begin() as conn:
    result = conn.execute(text("SELECT version();"))
    print(result.fetchone())
```

```python
sql_stmt = """DROP TABLE IF EXISTS public.land_tblnorthwind"""
with engine.begin() as conn:
    conn.execute(text(sql_stmt))
```

### get list of tables in your current database

```python
sql_stmt = f"""
SELECT *
FROM information_schema.tables
"""

with engine.connect() as conn:
    res = conn.execute(text(sql_stmt)).fetchall()
    
print(res)
```

```python
sql_stmt = f"""
SELECT table_name 
FROM information_schema.tables 
WHERE table_schema='public';
"""

with engine.connect() as conn:
    res = conn.execute(text(sql_stmt)).fetchall()
    
print(res)
print(res[0][0])
```

```python
sql_stmt = f"""
SELECT column_name, data_type 
FROM information_schema.columns 
WHERE table_name ='{NAME_TBL_NORTHWIND}'
ORDER BY ordinal_position
"""

with engine.connect() as conn:
    res = conn.execute(text(sql_stmt)).fetchall()
    
print(res)
```

```python
sql_stmt = f"""
SELECT constraint_name, constraint_type 
FROM information_schema.table_constraints 
WHERE table_name ='{NAME_TBL_NORTHWIND}'
"""

with engine.connect() as conn:
    res = conn.execute(text(sql_stmt)).fetchall()
    
print(res)
```

# Scrape manually downloaded page

```python
with open("IMDb_chart_top.html", "r", encoding="utf-8") as f:
    html = f.read()
print(html[:500])
```

```python
from bs4 import BeautifulSoup  # Import BeautifulSoup to parse HTML

soup = BeautifulSoup(html, "html.parser")  # Parse the HTML content of the page
```

```python
from pprint import pprint 

rows = soup.find_all("h3", class_="ipc-title__text") 
pprint(rows[:5])
```

```python
rows = soup.select("li.ipc-metadata-list-summary-item")  # uses CSS selector to find all <li> elements with class "ipc-metadata-list-summary-item"
pprint(rows[:5])
```

```python
rows = soup.find_all("li", class_="ipc-metadata-list-summary-item")  # same as above but pythonic
pprint(rows[:5])
first_25_movies = rows[:25]
```

```python
for m in first_25_movies:
    title = m.find("h3", class_="ipc-title__text")
    print(title.text, end=" - ")
    rating = m.find("span", class_="ipc-rating-star--rating")
    print(rating.text)
```

```python
import pandas as pd
titles = [m.find("h3", class_="ipc-title__text").text for m in first_25_movies]
ratings = [float(m.find("span", class_="ipc-rating-star--rating").text.replace(',', '.')) for m in first_25_movies]
df = pd.DataFrame({"title": titles, "rating": ratings})
df.head()
```

```python
df.to_csv("top_25_movies.csv", index=False)
```

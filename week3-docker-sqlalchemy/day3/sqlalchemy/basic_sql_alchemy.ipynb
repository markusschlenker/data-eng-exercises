```python
"""
SQLAlchemy ORM Example with PostgreSQL

This script demonstrates:
1. Connecting to PostgreSQL using SQLAlchemy ORM
2. Defining a mapped class (ORM model)
3. Creating tables in the database
4. Inserting data
5. Querying data using ORM session
6. Executing raw SQL queries with text()
"""

# 1️⃣ Import required libraries
from sqlalchemy import create_engine, Column, Integer, String, text
from sqlalchemy.orm import declarative_base, sessionmaker

# 2️⃣ Connect to PostgreSQL
# Replace with your credentials: username, password, host, port, database_name
DATABASE_URL = "postgresql://markus:1markus!@localhost:5432/postgres"


# create_engine manages the connection pool and SQL execution
engine = create_engine(DATABASE_URL, echo=True)  # echo=True prints all SQL statements

# 3️⃣ Define a Base class for ORM models
# All ORM models will inherit from this base
Base = declarative_base()

# 4️⃣ Define the ORM model (mapped class)
# Each class represents a table in the database
NAME_TBL_USERS = "users_ex2_sqlalchemy_orm"  # table name in PostgreSQL
class User(Base):
    __tablename__ = NAME_TBL_USERS  # table name in PostgreSQL
    __table_args__ = {"schema": "public"}  # optional schema

    # Columns
    id = Column(Integer, primary_key=True)  # primary key, auto-increment
    name = Column(String, nullable=False)   # Name column, required
    email = Column(String, nullable=True)   # Email column, optional

    def __repr__(self):
        # Optional: representation of object for debugging
        return f"<User(id={self.id}, name='{self.name}', email='{self.email}')>"

# 5️⃣ Create tables in the database
# create_all() only creates tables that do not already exist
Base.metadata.create_all(engine)
print(f"[INFO] Table '{NAME_TBL_USERS}' created or already exists.")

# 6️⃣ Create a session factory
# sessionmaker binds the engine to a session object
SessionLocal = sessionmaker(bind=engine)

# 7️⃣ Insert data using ORM session
# Sessions handle transactions automatically
session = SessionLocal()

# Create a new User object
new_user = User(name="Alice", email="alice@example.com")

# Add the object to the session
session.add(new_user)

# Commit the transaction to persist the data in the database
session.commit()
print(f"[INFO] Inserted one row into '{NAME_TBL_USERS}' table.")

# Optional: refresh the object to get autogenerated fields like id
session.refresh(new_user)
print(f"[INFO] New user ID: {new_user.id}")

# 8️⃣ Query data using ORM session
# Query all users
users = session.query(User).all()
print(f"[INFO] All users in the table '{NAME_TBL_USERS}':")
for user in users:
    print(user)

# Filter example: get a specific user by name
alice = session.query(User).filter(User.name == "Alice").first()
print(f"[INFO] Filtered user: {alice}")

# 9️⃣ Execute raw SQL using text()
# Sometimes you need raw SQL for advanced queries
result = session.execute(text("SELECT version();"))
version = result.fetchone()
print("[INFO] PostgreSQL version:", version[0])

# 1️⃣0️⃣ Close the session
session.close()
```

```python
"""
SQLAlchemy Core Example with PostgreSQL

This script demonstrates:
1. Connecting to PostgreSQL
2. Defining a table
3. Creating tables in the database
4. Inserting data
5. Querying data using both SQLAlchemy Core and raw SQL (text())
"""

# 1️⃣ Import required libraries
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String, insert, select, text

# 2️⃣ Connect to PostgreSQL
# Replace with your credentials: username, password, host, port, database_name
engine = create_engine(
    'postgresql://markus:1markus!@localhost:5432/postgres'
)
#dialect://username:password@hostname:portnumber/databasename
# The engine manages connections and sends SQL statements to the DB

# 3️⃣ Define a Table
# MetaData holds information about tables and schema
metadata = MetaData(schema="public")  # optional schema, default is 'public'

# Define a table called 'users' with three columns: id, name, email
NAME_TBL_USERS = "users_ex1_sqlalchemy_core"
users = Table(
    NAME_TBL_USERS,
    metadata,
    Column("id", Integer, primary_key=True),  # Primary key, auto-increments
    Column("name", String),                   # Name column (text)
    Column("email", String)                   # Email column (text)
)

# 4️⃣ Create Tables in the Database
# If the table already exists, create_all() does nothing
metadata.create_all(engine)
print(f"[INFO] Table '{NAME_TBL_USERS}' created or already exists.")

# 5️⃣ Insert Data into the Table
# Using SQLAlchemy Core 'insert' construct
with engine.connect() as conn:
    stmt = insert(users).values(name="Alice", email="alice@example.com")
    conn.execute(stmt)  # execute the INSERT statement
    conn.commit()       # commit the transaction
print(f"[INFO] Inserted one row into '{NAME_TBL_USERS}' table.")

# 6️⃣ Query Data using SQLAlchemy Core 'select'
with engine.connect() as conn:
    stmt = select(users)      # SELECT * FROM users
    results = conn.execute(stmt)
    print("[INFO] Query Results:")
    for row in results:
        print(row)

# 7️⃣ Query Data using raw SQL with 'text()'
with engine.connect() as conn:
    result = conn.execute(text("SELECT version();"))  # PostgreSQL version
    version = result.fetchone()
    print("[INFO] PostgreSQL version:", version[0])
```

## Task 3.

I've decided that this month I want to stay away from wines with relatively high alcohol content. To do that, I'm going to avoid any wines that have a greater alcohol content than the mean alcohol content, and you're going to help me do this. To achieve this, let's do the following: 

  * Find the mean alcohol content, separately, for reds and whites.  
  * Create a `Series` that holds whether each row in each `DataFrame` (the reds, and whites) has a higher alcohol content than the mean. 
  * Merge this `Series` onto the `DataFrame`. I can imagine doing this with either a `.join()` or using `pd.concat()`. For practice, do it with both. Note: merges with `Series` work the same way that they work with `DataFrames`.  
  * Return back to me all those rows that will help me stay away from those wines with a higher alcohol content.

```python
wine_df.columns
```

```python
wine_df.groupby('type').mean()['alcohol']
```

```python
red_mean_alc = redwine_df['alcohol'].mean()
white_mean_alc = whitewine_df['alcohol'].mean()
red_mean_alc
```

```python
red_alc_above_avg = (redwine_df['alcohol'] > red_mean_alc).rename('alcohol_above_avg')
white_alc_above_avg = (whitewine_df['alcohol'] > white_mean_alc).rename('alcohol_above_avg')
red_alc_above_avg
```

```python
pd.concat([redwine_df, red_alc_above_avg], axis=1)
```

```python
redwine_df = redwine_df.join(red_alc_above_avg)
whitewine_df = whitewine_df.join(white_alc_above_avg)
redwine_df
```

```python
redwine_df.describe()["alcohol"][["min","mean","max"]]
```

```python
redwine_df[redwine_df['alcohol_above_avg'] == False]
```

## side quest: 

In Pandas there is a deprecation warning on groupby with categorical data and implicit observed=False (default),
 observed=True will be the new default in Pandas 3


```python
# see notes in release notes for pandas 3.0.0 https://pandas.pydata.org/pandas-docs/stable/whatsnew/v3.0.0.html#improved-behavior-in-groupby-for-observed-false

df = pd.DataFrame(
    {
        "key1": pd.Categorical(list("aabb"), categories=list("abc")),
        "key2": [1, 1, 1, 2],
        "values": [1, 2, 3, 4],
    }
)
df
```

```python
gb = df.groupby("key1", observed=False)  # include also not observed categories, in this case 'c'
gb[["values"]].apply(lambda x: x.sum())
```

```python
gb = df.groupby("key1", observed=True)  # only include observed categories, in this case 'a' and 'b', 'c' is not included in the result
gb[["values"]].apply(lambda x: x.sum())
```

Let's say that I now want to know how much my decision to avoid those wines with higher `alcohol` content is going to limit the `quality` of wines that I can drink. To figure this out, I want to know a couple of things:

 Use a `pivot table` to solve this. 

```python
redwine_df.columns
```

```python
redwine_df[redwine_df['alcohol_above_avg'] == True].groupby('quality').mean()['alcohol']
```
